1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

1. Accuracy
Accuracy is most appropriate in situations where:

Both classes (positive and negative) are balanced, and
The cost of false positives and false negatives is similar.
Example Scenario: Predicting whether a machine in a factory is functioning or not.
Rationale: If the number of functioning and non-functioning machines is roughly equal, and the cost of misclassification is not heavily weighted toward one type of error, accuracy provides a straightforward and reliable evaluation of the model.

2. Sensitivity (True Positive Rate)
Sensitivity is most critical in situations where:

It is important to identify all positives, and
Missing a positive case has a high cost.
Example Scenario: Detecting a disease like cancer during medical screening.
Rationale: Missing a positive diagnosis (false negative) could have severe consequences for the patient’s health. Therefore, ensuring a high sensitivity ensures that most positive cases are caught.

3. Specificity (True Negative Rate)
Specificity is most important in scenarios where:

It is crucial to avoid false positives, and
Incorrectly labeling a negative as positive has a high cost.
Example Scenario: Screening for a rare disease before expensive or invasive follow-up tests.
Rationale: Since the disease is rare, a high specificity ensures that most healthy individuals are correctly identified, reducing unnecessary stress, costs, or invasive testing for false positives.

4. Precision (Positive Predictive Value)
Precision is valuable when:

The cost of false positives is higher than false negatives, and
You want to ensure that when a model predicts positive, it’s likely correct.
Scenario: Identifying defective products on an assembly line.
Rationale: In this case, predicting a product as defective when it is not (false positive) could lead to unnecessary rework or disposal, wasting time and resources. A high precision ensures that when a product is flagged as defective, it is very likely truly defective, reducing costly and unnecessary interventions.

Specificity is key when false positives are costly (e.g., rare disease screenings).
Precision is crucial when false positives must be minimized (e.g., fraud detection or identifying defective products).
Each metric’s relevance depends on the specific context and the cost of errors.

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>

But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. In linear regression, the coefficients represent the direct, linear relationship between each predictor variable and the predicted outcome, where a higher coefficient indicates a stronger influence of that feature on the target variable. In contrast, feature importances in decision trees indicate the relative contribution of each feature in reducing impurity (such as Gini or entropy) at each decision node, but they do not directly show how the feature values impact the outcome. Decision trees are more flexible in capturing non-linear relationships, whereas linear regression assumes a constant, additive effect from each predictor.
